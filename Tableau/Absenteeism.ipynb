Examinethe‘Absenteeism_data.csv’carefully.Then,usethefollowingasaguidetohowyoushouldpreparethedataforfurtheranalysis:•Drop the ‘ID’ column•Split the reasons for absence into multiple dummy variables, and then group them in the following way:  ➢Group 1: Columns 1 to 14➢Group 2: Columns 15, 16, and 17➢Group 3: Columns 18, 19, 20, and 21➢Group 4: Columns 22 to 28•After you’ve done that, don’t forget to drop the ‘Reason for Absence’ column.•Extract the month value and the day of the week from the ‘Date’ column. Then, drop the ‘Date’ column as well.•Turn the data from the ‘Education’ column into binary data, by mapping the value of 0 to the values of 1, and the value of 1 to the rest of the values found in this column.

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("Absenteeism_data.csv")
raw_data.head()
```

```python
#pd.options.display.max_columns or max rows

df = raw_data.copy()
df.columns
```

```python
df.info() #no non null
```

```python
df.drop(['ID'], axis = 1, inplace=True)
```

```python
print(f"Max:{df['Reason for Absence'].max()}, Min:{df['Reason for Absence'].min()}, Unique/missing:{sorted(df['Reason for Absence'].unique())}")
```

```python
## 20 is missing
reasons = pd.get_dummies(df['Reason for Absence'], drop_first=True) #drop first to remove multicollinarity
reasons
```

```python
reasons['check']=reasons.max(axis=1)
reasons['check'].unique() # no two resons were given
```

## grouping

 ➢Group 1: Columns 1 to 14➢Group 2: Columns 15, 16, and 17➢Group 3: Columns 18, 19, 20, and 21➢Group 4: Columns 22 to 28•

```python
r1=reasons.loc[:,1:14].max()
r2=reasons.loc[:,15:17].max()
r3=reasons.loc[:,18:21].max()
r4=reasons.loc[:,22:28].max()
```

```python
df_plus_reasons = pd.concat([df,r1,r2,r3,r4], axis=1)
df_plus_reasons.head()
```

```python
df_plus_reasons.tail()
```

```python
df_plus_reasons.info()
```

```python
df_plus_reasons.drop(['Reason for Absence'],axis=1, inplace=True)
```

```python
df_plus_reasons['Date'] = pd.to_datetime(df_plus_reasons['Date'])
```

```python
df_plus_reasons['Date_Month'] = df_plus_reasons['Date'].dt.month
```

```python
df_plus_reasons.drop(['Date'],axis=1, inplace=True)
```

```python
df_plus_reasons['Education_enhanced']=df_plus_reasons['Education'].map(lambda x: 0 if x==1 else 1)
df_plus_reasons['Education_enhanced'].unique()
```

```python
df_plus_reasons.head()
```

```python
df_plus_reasons.columns
```

```python
df_plus_reasons.rename(columns={
    0: 'reason_1',
    1: 'reason_2',
    2: 'reason_3',
    3: 'reason_4'
}, inplace=True)
df_plus_reasons.head()
```

```python
df_plus_reasons.to_csv('absenteeism_mod.csv', encoding = 'utf8')
```
